# 문자열(String)
글자들의 모음. 큰따옴표(" ") 또는 작은따옴표(' ')를 사용한다.

```python
# 생성 및 여러 줄 문자열
name = "홍길동"
poem = """나 보기가 역겨워
가실 때에는
죽어도 아니 눈물 흘리오리다
""" 
print(poem)

# 문자열 연산
greeting = "안녕"
print(greeting + " " + name)  # 이어붙이기
print(name * 3)  # 반복하기

# [주의] 숫자와 직접 더하기 불가
# print(name + 20)
print(name + str(20))  # 형변환 함수
```
